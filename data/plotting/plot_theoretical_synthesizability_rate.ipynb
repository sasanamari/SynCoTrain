```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
res = pd.read_csv('theoretical_synthesizability_rate.csv')
```

```python
res
```

```python
res = res[res['Theoretical Database']!='FTCP generated'] #not using FTCP results here.
```

```python
res.plot(x='Theoretical Database', y='Synthesizability (%)', kind='bar')
plt.xlabel('Theoretical Database')
plt.ylabel('Synthesizability (%)')
plt.title('Recall for Theoretical Databases')
plt.show()
```

```python
# import matplotlib.pyplot as plt

# Your original code with improvements:
res.plot(x='Theoretical Database', y='Synthesizability (%)', kind='bar', color='skyblue', edgecolor='black', legend=False)

plt.xlabel('Theoretical Database', fontsize=12, fontweight='bold')
plt.ylabel('Synthesizability (%)', fontsize=12, fontweight='bold')
plt.title('Predicted Positive Rate for Theoretical Databases', fontsize=14, fontweight='bold')

# Improve the tick labels' readability
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)

# Adding gridlines
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Setting a tight layout to ensure no clipping of labels
plt.tight_layout()

# Save the figure in high resolution suitable for publications
plt.savefig('theoretical_synth_barplot.png', dpi=300, format='png')

plt.show()
```
